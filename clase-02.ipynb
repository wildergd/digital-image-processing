```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
def imshow(image: np.ndarray, cmap = 'gray') -> None:
    plt.axis('off')
    plt.imshow(image, cmap)
```

```python
# cargar imágenes
image = cv.imread('./images/im3.jpeg', cv.IMREAD_GRAYSCALE)
imshow(image)
```

```python
image.shape
```

```python
x, y = np.mgrid[0:image.shape[0], 0:image.shape[1]]
fig1 = plt.figure()
ax = fig1.add_subplot(projection = '3d')
ax.plot_surface(x, y, image, cmap = plt.cm.hot)
plt.show()
```

```python
fig2 = plt.figure(figsize = (16, 4))
fig2.add_subplot(1, 2, 1)
plt.axis('off')
plt.imshow(image, cmap = 'gray', vmin=0, vmax=255)
fig2.add_subplot(1, 2, 2)
plt.hist(image.ravel(), 256, [0, 256])
plt.show()
```

```python
min = np.amin(image)
max = np.amax(image)

m = 255 / (max - min)
b = -m * min

print(m, b)

image_norm = (m * image + b).astype(np.uint8)
```

```python
fig3 = plt.figure(figsize = (16, 8))
fig3.add_subplot(2, 2, 1)
plt.axis('off')
plt.imshow(image, cmap = 'gray', vmin=0, vmax=255)
fig3.add_subplot(2, 2, 2)
plt.hist(image.ravel(), 256, [0, 256])
fig3.add_subplot(2, 2, 3)
plt.axis('off')
plt.imshow(image_norm, cmap = 'gray', vmin=0, vmax=255)
fig3.add_subplot(2, 2, 4)
plt.hist(image_norm.ravel(), 256, [0, 256])
plt.show()
```

```python
min = np.amin(image)
max = np.amax(image)

m = 255 / (max - min)
b = -m * min

print(m, b)

image_inverted = 255 - image
image_inverted_norm = (m * (255 - image) + b).astype(np.uint8)
```

```python
fig4 = plt.figure(figsize = (16, 8))
fig4.add_subplot(3, 2, 1)
plt.axis('off')
plt.imshow(image, cmap = 'gray', vmin=0, vmax=255)
fig4.add_subplot(3, 2, 2)
plt.hist(image.ravel(), 256, [0, 256])
fig4.add_subplot(3, 2, 3)
plt.axis('off')
plt.imshow(image_inverted, cmap = 'gray', vmin=0, vmax=255)
fig4.add_subplot(3, 2, 4)
plt.hist(image_inverted.ravel(), 256, [0, 256])
fig4.add_subplot(3, 2, 5)
plt.axis('off')
plt.imshow(image_inverted_norm, cmap = 'gray', vmin=0, vmax=255)
fig4.add_subplot(3, 2, 6)
plt.hist(image_inverted_norm.ravel(), 256, [0, 256])
plt.show()
```
