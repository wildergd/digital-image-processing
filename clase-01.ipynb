```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def imshow(image: np.ndarray, cmap = 'gray') -> None:
    plt.axis('off')
    plt.imshow(image, cmap)
```

```python
# cargar imgenes
image = cv.imread('./images/im1.jpg')
imshow(image)
```

```python
imshow(image[120:200, 70:150])
```
