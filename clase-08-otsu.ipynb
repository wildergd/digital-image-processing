```python
import sys
sys.dont_write_bytecode = True

import math
import numpy as np
import cv2 as cv
import lib.image_utils as ImgUtils
import matplotlib.pyplot as plt
%matplotlib inline

import contextlib
with contextlib.redirect_stdout(None):
    np.seterr(divide = 'ignore')

import warnings
warnings.filterwarnings('ignore')
```

```python
image = cv.imread('./images/im1.jpg', cv.IMREAD_GRAYSCALE)

threshold, ic_var = ImgUtils.otsu(image)
_, thresh1 = cv.threshold(image, threshold, 255, cv.THRESH_BINARY)

ImgUtils.imshow2(
    image,
    thresh1,
    titles = ['Original', f'Binarized (threshold = {threshold})'],
    titleColor = "white"
)
```

```python
# def bf_threshold(image):
vart = list([])
min = np.min(image)
max = np.max(image)
for t in range(min, max + 1, 1):
    c1 = image[image < t]
    c2 = image[image >= t]
    var1 = np.var(c1)
    var2 = np.var(c2)
    vart.append(var1 + var2)

plt.plot(np.arange(min, max + 1, 1), vart)
```
